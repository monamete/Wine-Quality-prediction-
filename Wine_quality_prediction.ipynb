**Feature Selection**

```python
#create the classification version of target variable
df["goodquality"] = [1 if x>=7 else 0 for x in df["quality"]]
x = df.drop(["quality","goodquality"],axis=1)
y= df["goodquality"]
```

**Splitting Dataset**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

**LogisticRegression:**

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

**Using KNN:**

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
y_pred = knn.predict(x_test)

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

**Using SVC**

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(x_train,y_train)
y_pred = svc.predict(x_test)

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

**Using Decision Tree:**

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred = dt.predict(x_test)

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

**Using Random Forest:**

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred = rf.predict(x_test)

from sklearn.metrics import accuracy_score
print((accuracy_score(y_test,y_pred))*100)
```

```python
##Hence I will use Random Forest algorithms for training my model.

```

**Cross validation**

```python
from sklearn.model_selection import KFold,StratifiedKFold,cross_val_score
cv_score = cross_val_score(rf,x,y,cv=10)
print(cv_score)
print(np.mean(cv_score))
```

```python
cv_score = cross_val_score(lr,x,y,cv=10)
print(cv_score)
print(np.mean(cv_score))
```

```python
cv_score = cross_val_score(svc,x,y,cv=10)
print(cv_score)
print(np.mean(cv_score))
```

```python
cv_score = cross_val_score(dt,x,y,cv=10)
print(cv_score)
print(np.mean(cv_score))
```

```python
cv_score = cross_val_score(knn,x,y,cv=10)
print(cv_score)
print(np.mean(cv_score))
```

**We have cross validated 5 algorithms,out of which Random Forest performed well with highest accuracy 87.55%**

**Hyperparameter Tuning**

```python
#lets create an object of GridSearchCV
from sklearn.model_selection import GridSearchCV

# Define the parameter grid for RandomForestClassifier
params = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}


out = GridSearchCV(estimator=rf,param_grid=params,cv=10)
out.fit(x_train,y_train)
```

```python
#train model using GridSearchCV object
out.fit(x,y)
```
